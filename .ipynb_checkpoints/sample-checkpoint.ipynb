```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
import pandas as pd
import os
directory = r'C:\Users\dkost\Desktop\Python\Poland_appartment\data'

filenames = [
    'apartments_pl_2023_08.csv',
    'apartments_pl_2023_09.csv',
    'apartments_pl_2023_10.csv',
    'apartments_pl_2023_11.csv',
    'apartments_pl_2023_12.csv',
    'apartments_pl_2024_01.csv',
    'apartments_pl_2024_02.csv',
    'apartments_pl_2024_03.csv',
    'apartments_pl_2024_04.csv',
    'apartments_pl_2024_05.csv',
    'apartments_pl_2024_06.csv'
]

dfs = [pd.read_csv(os.path.join(directory, filename)) for filename in filenames]
df = pd.concat(dfs, ignore_index=True)


combined_filepath = os.path.join(directory, 'combined_apartments_pl.csv')
df.to_csv(combined_filepath, index=False)

print(f"Combined file saved to: {combined_filepath}")
```

```python
df_sorted = df.sort_values(by='price', ascending=False)

sns.set_theme(style='darkgrid')  # Доступные стили: 'darkgrid', 'whitegrid', 'dark', 'white', 'ticks'
sns.set_palette('viridis')  # Примеры палитр: 'deep', 'muted', 'bright', 'dark', 'colorblind'

plt.figure(figsize=(18,8))
sns.barplot(data=df_sorted,x='city', y='price')
plt.xlabel('City')
plt.ylabel('Price in zł')
plt.title('Mean price of apartment in Polish cities \n 1e6 =1 000 000 ')
```

```python
plt.figure(figsize=(18,8))
sns.scatterplot(data=df_sorted,x='city', y='price', hue='buildYear')
plt.xlabel('City')
plt.ylabel('Price in zł')
plt.title('Mean price of apartment in Polish cities and build year \n 1e6 =1 000 000 ')
```

```python
plt.figure(figsize=(18,8))
sns.scatterplot(data=df_sorted, x='city', y='price', hue='centreDistance', palette='coolwarm')
plt.xlabel('City')
plt.ylabel('Price in zł')
plt.title('Mean price of apartment in Polish cities and distance to city centre (km) \n 1e6 =1 000 000 ')
```
